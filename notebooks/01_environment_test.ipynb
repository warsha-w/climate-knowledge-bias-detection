```python
import pandas as pd
import requests
import spacy

# Test 1: Check Data Science Libraries
print("System Check Complete.")
print("Pandas version:", pd.__version__)
print("Requests version:", requests.__version__)

# Test 2: Check the NLP AI Engine
print("\n--- AI NLP Engine Test ---")
nlp = spacy.load("en_core_web_sm")
text = "The University of Sri Jayewardenepura  is currently researching severe climate floods in Sri Lanka."
doc = nlp(text)

print("Sentence:", text)
print("Extracted Entities:")
for ent in doc.ents:
    print(f"- {ent.text} (Type: {ent.label_})")
```
